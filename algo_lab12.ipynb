```python
class GraphNode:
    
#     neighbors=[]
#     distance=0
    
    def __init__(self, id):
        self.id = id
#         self.edges = []
#         self.visited=False
        self.neighbors=[]
        self.distance=0
    
#     def add_neighbor(self,node):
#         self.neighbors.append(node)
        
    
    def add_neighbor(self, node, weight):
        self.neighbors.append([node,weight])
#         this->neighbors.add(node_id, weight)
    
    
    def get_neighbors(self): 
        return self.neighbors
```

```python
class Graph:
    
    def __init__(self,vertices): 
        self.V= vertices #No. of vertices 
        self.graph = [] # default dictionary  
                                # to store graph 
    
    def build_Graph(self,f2):
        

    #     initialize the graph array with GraphNode
        for i in range(0,self.V):
            newnode=GraphNode(i)
            self.graph.append(newnode)

        #     read from file and construct a graph

        for line in f2:
            id1=(line[0])
            id2=(line[1])
            weight=(line[2])


    #         GraphNode(id1).add_neighbor(GraphNode(id2))
            self.graph[id1].add_neighbor(self.graph[id2],weight)
       

        
        
    def find(self, parent, node_id): 
        if parent[node_id] == node_id: 
            return node_id 
        return self.find(parent, parent[node_id]) 
    
    def union(self, parent, rank, x, y): 
        xroot = self.find(parent, x) 
        yroot = self.find(parent, y) 
  
        # Attach smaller rank tree under root of  
        # high rank tree (Union by Rank) 
        if rank[xroot] < rank[yroot]: 
            parent[xroot] = yroot 
        elif rank[xroot] > rank[yroot]: 
            parent[yroot] = xroot 
  
        # If ranks are same, then make one as root  
        # and increment its rank by one 
        else : 
            parent[yroot] = xroot 
            rank[xroot] += 1
            
    def Krus(self):
#         Read in the file and put edges in acsending order
        edges=[]
        for i in self.graph:
            for k in i.neighbors:
                edges.append([i.id,k[0].id,k[1]])
    
        sorted_edges=[]
        
        sorted_edges =  sorted(edges,key=lambda item: item[2])

        parent = []
        rank = [] 
  
        # Create V subsets with single elements 
        for node in range(len(self.graph)): 
            parent.append(node) 
            rank.append(0)
    
    
        mst_weight = 0
       
    #       Define your own class for edge information
        for e in sorted_edges:
            node1id = e[0]
            node2id = e[1]
            
            x = self.find(parent, node1id) 
            y = self.find(parent ,node2id) 
            
            if x!=y:
#                 union the two nodes
                self.union(parent, rank, x, y)
                mst_weight += e[2]
        
        print("The total weight for the MST is ", mst_weight)
       
    
    
    
```
