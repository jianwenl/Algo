```python
def mergeSort(a, aux, low, high):
        if high - low <= 1:
            return a
        mid = (low + high)//2
        mergeSort(a, aux, low, mid)
        mergeSort(a, aux, mid, high)
        merge(a, aux, low, mid, high)
        return a

def merge(a, aux, low, mid, high):
        #copy a[low:high] to aux[low:high] 
        aux[low:high] = a[low:high]
        #merge from aux to a
        i = low
        j = mid
        k = low
        while i < mid and j < high:
    
            if aux[i] <= aux[j]:
                a[k] = aux[i]
                i+=1
            else:                   
                a[k] = aux[j]
                j+=1
            k+=1
        
        while(i < mid):
            a[k] = aux[i]
            k+=1
            i+=1
        while(j < high):
            a[k] = aux[j]
            k+=1
            j+=1
```
