```python
class BinaryHeap:
    _size = 0
    heap = []
    
    def __init__(self, A, n):
        self._size = n
        self.heap = A
        # 初始化siftdown
        for i in range(self._size-1,0,-1):
            self.siftDown(i)
    
    def size(self):
        return self._size
    
    def swap(self, i, j):
        self.heap[i], self.heap[j] = self.heap[j], self.heap[i]
        
    def siftDown(self, index):
        if index >= self._size:
            return
        left = index * 2
        right = left +1

        if left <= self._size and self.heap[index-1] < self.heap[left-1]:
            self.swap(left-1, index-1)
            self.siftDown(left)

        if right <= self._size and self.heap[index-1] < self.heap[right-1]:
            self.swap(right-1, index-1)
            self.siftDown(right)

    def delMax(self):
        res = self.heap[0]
        self.swap(0, (self._size - 1))
        self._size -= 1
        self.siftDown(1)
        return res
```

```python
def main_f(A):   
#     Create binary heap by the input array A   
    heap=BinaryHeap(A,len(A))
    for i in range(20):      
        outputfile=heap.delMax()
        print (outputfile)
```
