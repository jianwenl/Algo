```python
import random
def generateRandomNumbers(n):
    a=[]
    for i in range(n):
        b = random.randint(0, n)/n
        a.append(b)
    return a

generateRandomNumbers(100)
```

```python
def quickSort(alist):
    quickSortHelper(alist,0,len(alist)-1)

def quickSortHelper(alist,first,last):
    if first<last:

        #choose a random pivot
        index = random.randint(first, last-1)
        temp = alist[first]
        alist[first] = alist[index]
        alist[index] = temp
        
        splitpoint = partition(alist,first,last)

        quickSortHelper(alist,first,splitpoint-1)
        quickSortHelper(alist,splitpoint+1,last)        


def partition(alist,first,last):
    pivotvalue = alist[first]

    leftmarki = first+1
    rightmarkj = last

    done = False
    while not done:

        while leftmarki <= rightmarkj and alist[leftmarki] <= pivotvalue:
            leftmarki = leftmarki + 1

        while alist[rightmarkj] >= pivotvalue and rightmarkj >= leftmarki:
            rightmarkj = rightmarkj -1

        if rightmarkj < leftmarki:
            done = True
        else:
            temp = alist[leftmarki]
            alist[leftmarki] = alist[rightmarkj]
            alist[rightmarkj] = temp

    #swap
    temp = alist[first]
    alist[first] = alist[rightmarkj]
    alist[rightmarkj] = temp


    return rightmarkj
```
