```python
# class treeNode:
#     def __init__(self, val):
#         self.l_child = None
#         self.r_child = None
#         self.data = val

class Node:
    rc,lc,parent,value = None,None,None,0    

    def __init__(self, val):
        self.parent=None
        self.lc = None
        self.rc = None
        self.value = val
```

```python
class BST:
    data in BST
    root,_size = None,0

#     Constructor
    def __init__(self):
        self.root=None
        self._size = 0

    def inOrder(self,someNode):
        if someNode is None:
            pass
        else:
            self.inOrder(someNode.lc)
            print (someNode.value)
            self.inOrder(someNode.rc)
    
#     def inOrder(self):
#         if self.lc != None:
#             self.lc.inOrder()
#         print (self.value),
#         if self.rc != None:
#             self.rc.inOrder()

#      Read-only operation
#     search(val)
    def search(self,someNode,val):
#     // Base case
        if someNode is None:
            return False

#         // Hit
        if someNode.value == val:
            return True

#         // Go to left subtree
        elif someNode.value > val: 
            return self.search(someNode.lc, val)

#         Go to right subtree
        else:
            return self.search(someNode.rc, val)


#     Write operation
#     insert(tn, val)
    def insert(self,val):
        self._size = self._size+1
        if self.root is None:
            self.root = Node(val)
        else:
            self.insertWithNode(self.root, val)
    
    
    def insertWithNode(self,node,val):
        if node.lc is None and node.rc is None:#external node
            if val > node.value:
                newNode = Node(val)
                node.rc = newNode
                newNode.parent = node
            else:
                newNode = Node(val)
                node.lc = newNode
                newNode.parent = node
        else: #not external
            if val > node.value:
                if node.rc is not None:
                    self.insertWithNode(node.rc, val)
                else: #if empty node
                    newNode = Node(val)
                    node.rc = newNode
                    newNode.parent = node 
            else:
                if node.lc is not None:
                    self.insertWithNode(node.lc, val)
                else:
                    newNode = Node(val)
                    node.lc = newNode
                    newNode.parent = node 
    
    
#     def insert(self,val):
#         if _size == 0:
#             root.value = val
# #         if self.data == None:
# #             self.data = BST(val)
#         if val > self.val:
#             if self.rc == None:
#                 self.rc = BST(val)
#             else:
#                 self.rc.insert(val)
#         else:
#             if self.lc == None:
#                 self.lc = BST(val)
#             else:
#                 self.lc.insert(val)
```
