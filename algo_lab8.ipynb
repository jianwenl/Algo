```python
import random
```

```python
class Node:
    
    rc,lc,parent,value,height = None,None,None,0,0

    
    def __init__(self, data, height):
        self.parent=None
        self.lc = None
        self.rc = None
        self.value = data
        self.height = height
        self.priority = random.random()#double
```

```python
class Treap:
#     SAME data as in BST
    #_size,root=0,None
    
    

#     SAME Constructor as BST
    def __init__(self):
        self.root=None
        self._size = 0

#     SAME Tree traversal
    def inOrder(self,someNode):
        if someNode is None:
            pass
        else:
            self.inOrder(someNode.lc)
            print (someNode.value)
            self.inOrder(someNode.rc)

#     // Read-only operation
#     TODO: change return type
#     treeNode search(int val)
    
    
    def search(self,someNode,val):
#     // Base case
        if someNode is None:
            return None

#         // Hit
        if someNode.value == val:
            return someNode

#         // Go to left subtree
        elif someNode.value > val: 
            return self.search(someNode.lc, val)

#         Go to right subtree
        else:
            return self.search(someNode.rc, val)
    
    
    

#     // Write operation
#     TODO: adjust height during insertion
#     int insert(treeNode tn, int val)
    
    def insert(self,val):
        self._size = self._size+1
        if self.root is None:
            self.root = Node(val,0)
        else:
            self.insertWithNode(self.root, val)
    
    
    def insertWithNode(self,node,val):
        if node.lc is None and node.rc is None:#external node
            if val > node.value:
                newNode = Node(val,0)
                node.rc = newNode
                newNode.parent = node
                
                self.updateHeight(newNode.parent)
                self.siftUp(newNode)
            else:
                newNode = Node(val,0)
                node.lc = newNode
                newNode.parent = node
                self.updateHeight(newNode.parent)
                self.siftUp(newNode)
        else: #not external
            if val > node.value:
                if node.rc is not None:
                    self.insertWithNode(node.rc, val)
                else: #if empty node
                    newNode = Node(val,0)
                    node.rc = newNode
                    newNode.parent = node
                    self.updateHeight(newNode.parent)
                    self.siftUp(newNode)
            else:
                if node.lc is not None:
                    self.insertWithNode(node.lc, val)
                else:
                    newNode = Node(val,0)
                    node.lc = newNode
                    newNode.parent = node
                    self.updateHeight(newNode.parent)
                    self.siftUp(newNode)
        
            
    def siftUp(self,node):
        while (node.parent!=None and node.parent.priority<node.priority):
            self.rotate(node)
    
    

#     // Rotation
#     void rotate(treeNode tn)
#     void rotate_left(treeNode tn)
#     void rotate_right(treeNode tn)
    
    
    def rotate(self,someNode):
#     TODO: base case

        if not someNode.parent:
            return
        elif someNode is someNode.parent.rc:
            self.rotate_left(someNode)
        else: 
            self.rotate_right(someNode)
    
    def rotate_right(self,x):
        y = x.parent
        z = x.rc
        yParent = y.parent

#         // Update y's parent's child
#         // Need to decide left or right child
        if yParent == None:
            self.root = x
        elif yParent.lc == y:
            yParent.lc = x
        else:
            yParent.rc = x

#         // work on x
        x.parent = yParent
        x.rc = y

#         // work on y
        y.parent = x
        y.lc = z

#         // work on z
        if z != None:
            z.parent = y

        self.updateHeight(y)
        self.updateHeight(x)
        
    def rotate_left(self,y):
        x = y.parent
        z = y.lc
        xParent = x.parent

#         // Update y's parent's child
#         // Need to decide left or right child
        if xParent == None:
            self.root = y
        elif xParent.rc == x:
            xParent.rc = y
        else:
            xParent.lc = y

#         // work on y
        y.parent = xParent
        y.lc = x

#         // work on x
        x.parent = y
        x.rc = z

#         // work on z
        if z != None:
            z.parent = x

        self.updateHeight(x)
        self.updateHeight(y)



#     // Height adjustment
#     void updateHeight(treeNode tn)
    
    
    def updateHeight(self,someNode):
#         // base case
        if not someNode:
            return

        oldHeight = someNode.height
        if (someNode.lc != None and someNode.rc != None):
            newHeight = 1 + max(someNode.lc.height, someNode.rc.height)
        elif someNode.lc != None:
            newHeight = 1 + someNode.lc.height
        elif someNode.rc != None:
            newHeight = 1 + someNode.rc.height
        else:
            newHeight=0

        if oldHeight == newHeight:
            return    
#     // No need to go further
        else:
            someNode.height = newHeight
            self.updateHeight(someNode.parent)
    
```
